# Problem 2: Find Angle MBC

image.png

```python
import math

ab = float(input())
bc = float(input())
ac = math.sqrt(ab**2 + bc**2)
bm = ac/2.0
mc = bm

b = mc
c = bm
a = bc

angle1 = math.acos(a/(2*b))

angle1_degree = int(round((180 * angle1) / math.pi))
print(angle1_degree,"\u00B0",sep='')
```
