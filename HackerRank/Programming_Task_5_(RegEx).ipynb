# Problem 1: Detect Floating Point Number


image.png

```python
import re

pattern = r'^[+-.]\d+\.\d+|\d+\.\d+$|[+-]\.\d+'

n = int(input())

for i in range(n):
    target = input()
    if re.fullmatch(pattern,target):
        print('True')
    else:
        print('False')
```

# Problem 6: Regex Substitution

image.png

```python
import re
def subst(match):

    if match.group(0)=='&& ':
        return 'and '
    if match.group(0)=='|| ':
        return 'or '
    raise ValueError


n=int(input())
for _ in range(n):
    line=input()
    print(re.sub(r'((?<= )\&\& )|((?<= )\|\| )', subst, line))
```

# Problem 11: HTML Parser - Part 1

image.png

```python
from html.parser import HTMLParser

def show(attrs):
    for attr in attrs:
            print("-> {0} > {1}".format(attr[0], attr[1]))
            
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print(f"Start : {tag}")
        show(attrs)
            
    def handle_endtag(self, tag):
        print("End   : " + tag)

    def handle_startendtag(self, tag, attrs):
        print("Empty :", tag)
        show(attrs)
        
parser = MyHTMLParser()
parser.feed(''.join([input() for _ in range(int(input()))]))
```

# Problem 12: HTML Parser - Part 2

image.png

```python
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_comment(self, data):
        data = str(data).split('\n')
        if len(data)>1:
            print('>>> Multi-line Comment', *data, sep='\n')
        else:
            print('>>> Single-line Comment', data[0], sep='\n')            
    def handle_data(self, data):
        data = str(data).replace('\n', '')
        if len(data)>0:
            print('>>> Data', data, sep='\n')

html = ""       
for _ in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

# Problem 13: Detect HTML Tags, Attributes and Attribute Values

image.png

```python
from html.parser import HTMLParser

class CustomHTMLParser(HTMLParser):
    
    def custom_print_attrs(self, attrs):
        if attrs:
            for attr in attrs:
                print('->', attr[0], '>', attr[1])
    
    def handle_starttag(self, tag, attrs):
        print(tag)
        self.custom_print_attrs(attrs)

    def handle_startendtag(self, tag, attrs):
        print(tag)
        self.custom_print_attrs(attrs)

html_parser = CustomHTMLParser()
pages = [input() for i in range(int(input()))]

for page in pages:
    html_parser.feed(page)
```

# Problem 14: Validating UID

image.png

```python
import re

def is_valid_uid(uid):
    return all([
        re.search(r'[A-Z].*[A-Z]', uid),
        re.search(r'\d.*\d.*\d', uid),
        not re.search(r'\W', uid),
        len(set(uid)) == 10,
        len(uid) == 10
    ])

t = int(input().strip())

for i in range(t):
    uid = input().strip()
    if is_valid_uid(uid):
        print("Valid")
    else:
        print("Invalid")
```

# Problem 15: Validating Credit Card Numbers

image.png

```python
import re

for i in range(int(input())):
    
    text = input()
    pattern = re.compile(r"^([456][\d+]{3})-([\d+]{4})-([\d+]{4})-([\d+]{4})$|^([456][\d+]{15})$")

    if pattern.match(text):
        text = ''.join([i for i in text if i != '-'])
        if all([False if i*4 in text  else True for i in set(text)]):
            print("Valid")
        else:
            print("Invalid")
    else:
        print("Invalid") 
```
